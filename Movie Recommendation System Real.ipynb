```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):  # input text should be in string format
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)   # convert into string 
```
