## Under-Sampling

```python
Build a sample dataset containing similiar distribution of normal tracsection and fraudulent trasection
```

```python
Number of Fraud Trasection --> 492
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenate two dataframe
# axis 0 for all the fraud values will be added below legit.sample row bise
# axis 1 for will the column bise

new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
x=new_dataset.drop(columns = 'Class', axis = 1)
y=new_dataset['Class']
```

```python
x
```

```python
y
```

```python
Split thw data into Training and Testing Data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify = y, random_state = 2)
```

## Model Training

### Logistic Regression

```python
# Trianig the Logistic Regression Model with Training Data
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

### Model Evaluation

#### Accurace Score

```python
# Accuracy on Training Data
x_train_predition = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predition , y_train)
```

```python
print('Accuracy on Training Data:' ,training_data_accuracy)
```

```python
# Accuracy score on Testing data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on Testing Data:' , test_data_accuracy)
```

```python
# Accurcy score of Training data and Testing data is similiar
# If there is lots of diff in Training data and Testing Data than it is called over fitted and under fitted.
# Model is Over fitted if Training accuracy is 92% and the Testing accuracy is 50%
```
